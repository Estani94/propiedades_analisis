```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.remax.com.ar/listings/buy?page=0&pageSize=24&sort=%2BpriceUsd&in:operationId=1&in:eStageId=0,1,2,3,4&locations=in:::458@%3Cb%3ESanta%3C%2Fb%3E%20Fe%20Capital::::&landingPath=&filterCount=0&viewMode=listViewMode"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, 'html.parser')

# Extraer elementos
prices = [p.get_text(strip=True).replace("USD", "").strip()
          for p in soup.select('.card__price')]
addresses = [a.get_text(strip=True) for a in soup.select('.card__address')]
surfaces = []

# Extraer solo los m² terreno (filtrando por texto que contenga "m² terreno")
for tag in soup.find_all('span'):
    if tag.next_sibling and "m² terreno" in tag.next_sibling:
        surfaces.append(tag.get_text(strip=True))

# Asegurar que las listas tengan la misma longitud
data = list(zip(addresses, prices, surfaces))

# Crear DataFrame
df = pd.DataFrame(data, columns=["Dirección", "Precio USD", "Terreno m2"])

# Guardar en CSV
df.to_csv("propiedades.csv", index=False, encoding='utf-8')

# Leer CSV como DataFrame
df_leido = pd.read_csv("propiedades.csv")

print(df_leido)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Base de la URL, sin la parte de la página
base_url = "https://www.remax.com.ar/listings/buy?page={page}&pageSize=24&sort=%2BpriceUsd&in:operationId=1&in:eStageId=0,1,2,3,4&locations=in:::458@%3Cb%3ESanta%3C%2Fb%3E%20Fe%20Capital::::&landingPath=&filterCount=0&viewMode=listViewMode"

headers = {
    "User-Agent": "Mozilla/5.0"
}

# Lista donde guardaremos los resultados
all_data = []

# Empezamos con la página 0
page_number = 0

while True:
    # Construir la URL con el número de página
    url = base_url.format(page=page_number)

    # Realizar la solicitud GET
    response = requests.get(url, headers=headers)

    # Si no hay contenido, terminamos
    if response.status_code != 200 or "No se encontraron propiedades" in response.text:
        print("No más páginas o error en la solicitud.")
        break

    # Parsear el HTML con BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extraer elementos
    prices = [p.get_text(strip=True).replace("USD", "").strip() for p in soup.select('.card__price')]
    addresses = [a.get_text(strip=True) for a in soup.select('.card__address')]
    surfaces = []

    # Extraer solo los m² terreno (filtrando por texto que contenga "m² terreno")
    for tag in soup.find_all('span'):
        if tag.next_sibling and "m² terreno" in tag.next_sibling:
            surfaces.append(tag.get_text(strip=True))

    # Si no hay datos, terminamos
    if not prices or not addresses or not surfaces:
        print("No se encontraron propiedades en esta página.")
        break

    # Asegurar que las listas tengan la misma longitud
    data = list(zip(addresses, prices, surfaces))

    # Agregar los datos extraídos a la lista principal
    all_data.extend(data)

    # Pasar a la siguiente página
    page_number += 1
    print(f"Página {page_number} procesada.")

# Crear DataFrame
df = pd.DataFrame(all_data, columns=["Dirección", "Precio USD", "Terreno m2"])

# Guardar en CSV
df.to_csv("propiedades.csv", index=False, encoding='utf-8')

# Leer CSV como DataFrame
df_leido = pd.read_csv("propiedades.csv")

print(df_leido)
```

```python
import pandas as pd

# Leer el CSV
df = pd.read_csv("propiedades.csv")

# Eliminar filas donde 'Precio USD' contenga 'ARS' o valores no numéricos como 'Solicitar precio'
df = df[~df['Precio USD'].str.contains('ARS', na=False)]
df = df[~df['Precio USD'].str.contains('Solicitar precio', na=False)]

# Eliminar puntos de los precios y convertir a número
df['Precio USD'] = df['Precio USD'].str.replace('.', '', regex=False)

# Convertir la columna 'Precio USD' a números, cualquier valor no numérico se convertirá a NaN
df['Precio USD'] = pd.to_numeric(df['Precio USD'], errors='coerce')

# Convertir la columna 'Terreno m2' a número (puede haber valores no numéricos, como texto o espacios, que hay que limpiar)
df['Terreno m2'] = pd.to_numeric(df['Terreno m2'], errors='coerce')

# Eliminar filas con valores nulos (NaN) en 'Precio USD' o 'Terreno m2'
df = df.dropna(subset=['Precio USD', 'Terreno m2'])

# Ver el DataFrame resultante
print(df)

# Guardar el DataFrame limpio de nuevo a CSV
df.to_csv("propiedades_limpias.csv", index=False, encoding='utf-8')
```

```python
# Filtrar las propiedades con un precio mayor a 30,000 USD
df_filtrado = df[df['Precio USD'] > 30000]

# Calcular el precio por m² para las propiedades filtradas
df_filtrado['Precio por m²'] = df_filtrado['Precio USD'] / df_filtrado['Terreno m2']

# Calcular el precio promedio del m² para las propiedades filtradas
precio_promedio_m2 = df_filtrado['Precio por m²'].mean()

# Mostrar el resultado
print(f"El precio promedio del m² para propiedades mayores a 30,000 USD es: {precio_promedio_m2:.2f} USD/m²")
```

```python
import re

# Función para separar número y texto
def separar_numero_letras(direccion):
    # Buscar la última secuencia de dígitos al final de la cadena
    match = re.search(r'(\D+)(\d+)$', direccion)
    if match:
        # Devuelve el texto (letras) y el número por separado
        return match.group(1).strip(), int(match.group(2))
    else:
        return direccion, None

# Aplicar la función a la columna 'Dirección'
df[['Dirección', 'Número']] = df['Dirección'].apply(lambda x: pd.Series(separar_numero_letras(x)))

# Ver el DataFrame resultante
print(df)
```

```python
# Calcular el precio por m² para cada propiedad
df['Precio por m²'] = (df['Precio USD'] / df['Terreno m2']).copy()

# Agrupar por Dirección y calcular el promedio del precio por m²
df_promedio = df.groupby('Dirección')['Precio por m²'].mean().reset_index()

# Mostrar el DataFrame con el promedio del precio por m² por Dirección
df_promedio.sort_values('Precio por m²', ascending=False, inplace=True)

df_promedio.head(20)
```

```python
# Agrupar por dirección: promedio del precio por m² y cantidad de propiedades
df_resumen = df.groupby('Dirección').agg({
    'Dirección': 'count'
}).rename(columns={'Dirección': 'Cantidad de propiedades'}).reset_index()

# Ordenar por precio por m² descendente
df_resumen.sort_values('Cantidad de propiedades', ascending=False, inplace=True)

# Mostrar las 20 direcciones más caras por m²
df_resumen.head(20)
```
